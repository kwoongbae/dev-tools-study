### 01. 라이브러리 로드 및 데이터셋 생성

```python
import pandas as pd
import geopandas as gpd
import numpy as np
from keplergl import KeplerGl
```

```python
df = pd.DataFrame()
df["GENERATOR_ID"] = [1,2,3,4,5]
df["LATITUDE_DEGREES"] = np.array([35,36.62, 36.65, 37.01, 37.26])
df["LONGITUDE_DEGREES"] = np.array([127, 128.44, 128.48, 127.925, 127.11])
df["ALTITUDE_METERS"] = np.array()
df
```

### 02. 시각화에 적합한 형태로 변형

```python
gdf = gpd.GeoDataFrame(
    df,
    geometry = gpd.points_from_xy(df["LATITUDE_DEGREES"], df["LONGITUDE_DEGREES"])
)
gdf.crs = "EPSG:4326"
gdf
```

```python
gdf_meter = gdf.to_crs("EPSG:5179")

gdf_meter["LONGITUDE_METERS"] = gdf_meter.geometry.x
gdf_meter["LATITUDE_METERS"] = gdf_meter.geometry.y
gdf_meter
```

### 03. Kepler로 시각화

```python
map_ = KeplerGl(height = 500)
map_.add_data(data=gdf_meter)
map_
```
